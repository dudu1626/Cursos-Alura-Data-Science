## Ajuste dos modelos Autocorrelacionados - AR

### Divis√£o dos dados para treino e teste
#### Ordem 1

```python
dados_ar1_p1 = dados_ar1[0:500]
dados_ar1_p2 = dados_ar1[500:600]
```

```python
modelo_ar1 = smtsa.ARMA(dados_ar1_p1['Preco'], order=(1, 0)).fit()
```

```python
modelo_previsto1 = modelo_ar1.predict(start=500,end=599)
```

```python
plt.figure(figsize=(12,8))
plt.plot(dados_ar1_p1['Data'],dados_ar1_p1['Preco'])
plt.plot(dados_ar1_p2['Data'],dados_ar1_p2['Preco'])
plt.plot(dados_ar1_p2['Data'], modelo_previsto1,'r.')
```

```python
modelo_previsto2 = modelo_ar2.predict(start=500,end=599)
```

```python
# somente o trecho a ser previsto:
plt.figure(figsize=(12,8))
plt.plot(dados_ar2_p2['Data'],dados_ar2_p2['Preco'])
plt.plot(dados_ar2_p2['Data'], modelo_previsto2,'r.')
```

## Auto ARIMA

```python
# !pip install pmdarima
```

```python
from pmdarima.arima import auto_arima
```

```python
modelo_busca = auto_arima(cresc_p1['Preco'].values,start_p=0, start_q=0, 
                            max_p=6, max_q=6, d=1,seasonal=False, trace=True,
                            error_action='ignore', suppress_warnings=True, stepwise=False)
```

### Nesse caso usamos o fator d (diferencia√ß√£o) fixo com o valor 1

```python
modelo_busca.aic()
# melhor valor de combina√ß√£o para os par√¢metros do order
```

### O procedimento de auto ARIMA j√° montou o modelo com a melhor combina√ß√£o dos par√¢metros order

```python
modelo_busca.fit(cresc_p1['Preco'].values)
```

```python
valores_preditos = modelo_busca.predict(n_periods=10)
```

```python
plt.plot(cresc_p2['Data'],valores_preditos, color='r')
plt.plot(cresc_p2['Data'],cresc_p2['Preco'], color='b')
```

## Modelos com Sazonalidade - SARIMA

```python
from pmdarima.arima.utils import nsdiffs
```

#### M√©todo para obten√ß√£o do valor do D para o modelo SARIMA

```python
D = nsdiffs(cresc_p1['Preco'].values, m=2, max_D=12, test='ch')
```

```python
D
```

#### Observar que o D est√° no modelo

```python
modelo_busca2 = auto_arima(cresc_p1['Preco'].values,start_p=0, start_q=0, max_p=6, max_q=6, d=1,D=1,
                            start_Q=1, start_P=1,
                            max_P=4, max_Q=4,m=2, seasonal=True, trace=True, error_action='ignore',
                            suppress_warnings=True, stepwise=False, maxiter=50)
```

### O modelo anterior sem a sazonalidade

```python
modelo_busca.aic()
```

### O novo modelo com a sazonalidade

```python
modelo_busca2.aic()
```

#### Quanto menor o valor, melhor

```python
modelo_busca2.fit(cresc_p1['Preco'].values)
```

```python
valores_preditos2 = modelo_busca2.predict(n_periods=10)
```

```python
plt.plot(valores_preditos2, color='r')
plt.plot(np.linspace(0,9,10),cresc_p2['Preco'], color='b')
```

## Acrescentado valores e retrinado o modelo

```python
from pmdarima.datasets import load_lynx
```

```python
dado_lynx = load_lynx()
```

```python
dado_lynx.shape
```

```python
from pmdarima import model_selection
```

```python
treino, teste = model_selection.train_test_split(dado_lynx, train_size=100)
```

```python
teste1 = teste[:10]
teste2 = teste[10:]
```

```python
modelo_arima = auto_arima(treino, start_p=1, start_q=1, d=0, max_p=5, max_q=5,
                          suppress_warnings=True, stepwise=True, error_action='ignore')
```

```python
previsoes = modelo_arima.predict(n_periods=teste1.shape[0])
```

```python
plt.plot(np.arange(0,treino.shape[0]),treino)
plt.plot(np.arange(treino.shape[0],treino.shape[0] + teste1.shape[0]),teste1)
plt.plot(np.arange(treino.shape[0],treino.shape[0] + teste1.shape[0]),previsoes)
```

## Retreinando

```python
modelo_arima.update(teste1)
```

```python
novas_pred = modelo_arima.predict(n_periods=teste2.shape[0])
```

```python
plt.plot(np.arange(0, treino.shape[0]), treino)
plt.plot(np.arange(treino.shape[0], treino.shape[0] + teste1.shape[0]), teste1)
plt.plot(np.arange(treino.shape[0] + teste1.shape[0],treino.shape[0] + teste1.shape[0] + teste2.shape[0]), teste2)
plt.plot(np.arange(treino.shape[0] + teste1.shape[0],treino.shape[0] + teste1.shape[0] + teste2.shape[0]), novas_pred)
```
