# Manipulando os dados

## Criando colunas numéricas

```python
# voltando a usar os dados sem filtragem
dados
```

```python
# criação de uma coluna com o custo do imóvel por mês
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
# criação de uma coluna com o custo do imóvel no ano
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype('str') + ' quarto(s) ' + ' e ' + \
                     dados['Vagas'].astype('str') + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')
dados.head()
```
