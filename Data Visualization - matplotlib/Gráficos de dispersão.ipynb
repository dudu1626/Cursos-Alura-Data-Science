```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
# definição geral do ambiente gráfico
fig = plt.figure(figsize=(16,9))
eixo = fig.add_axes([0,0,1,1])

# gráfico de dispersão
eixo.scatter(df.comprimento_sépala, df.largura_sépala)
eixo.set_title('Gráfio de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)
```

```python
# definição geral do ambiente gráfico
fig = plt.figure(figsize=(16,9))
eixo = fig.add_axes([0,0,1,1])

# gráfico de dispersão
eixo.scatter(df.comprimento_sépala, df.largura_sépala)
eixo.set_title('Gráfio de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)
eixo.tick_params(labelsize=15)
```

```python
# definição geral do ambiente gráfico
fig = plt.figure(figsize=(16,9))
eixo = fig.add_axes([0,0,1,1])

# definir cores distintas para cada tipo de Íris
cores = {'Iris-setosa' : 'r', 'Iris-versicolor' : 'b', 'Iris-virginica' : 'g'}

# para cada tipo de Íris, um gráfico de dispersão
for especie in df.espécie.unique():
    tmp = df[df.espécie == especie]
    eixo.scatter(tmp.comprimento_sépala, tmp.largura_sépala, color= cores[especie])

# configuração do gráfico
eixo.set_title('Gráfio de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)
eixo.tick_params(labelsize=15)
eixo.legend(cores, fontsize=20)
```

```python
# definição geral do ambiente gráfico
fig = plt.figure(figsize=(16,9))
eixo = fig.add_axes([0,0,1,1])

# definir cores distintas para cada tipo de Íris
cores = {'Iris-setosa' : 'r', 'Iris-versicolor' : 'b', 'Iris-virginica' : 'g'}
marcadores = {'Iris-setosa' : 'x', 'Iris-versicolor' : 'o', 'Iris-virginica' : 'v'}

# para cada tipo de Íris, um gráfico de dispersão
for especie in df.espécie.unique():
    tmp = df[df.espécie == especie]
    eixo.scatter(tmp.comprimento_sépala, tmp.largura_sépala, color= cores[especie], marker= marcadores[especie])

# configuração do gráfico
eixo.set_title('Gráfio de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)
eixo.tick_params(labelsize=15)
eixo.legend(cores, fontsize=20)
```

```python
# definição geral do ambiente gráfico
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

# média e desvio padrão
media = df.comprimento_pétala.mean()
desvio_padrao = df.comprimento_pétala.std()

# histograma
eixo.hist(df.comprimento_pétala, bins=20) # resultado normalizado
eixo.set_title("Histograma", fontsize=20, pad=10)
eixo.set_xlabel("Comprimento da pétala", fontsize=15)
eixo.grid(True)

# colocar média e desvio padrão
eixo.annotate('$media = {0:.2f}$ \n $desvio_padrao = {1:.2f}$'.format(media, desvio_padrao),
             xy=(4.5,20))
```

```python
# definição geral do ambiente gráfico
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

# média e desvio padrão
media = df.comprimento_pétala.mean()
desvio_padrao = df.comprimento_pétala.std()
mediana = df.comprimento_pétala.median()

# histograma
eixo.hist(df.comprimento_pétala, bins=20) # resultado normalizado
eixo.set_title("Histograma", fontsize=20, pad=10)
eixo.set_xlabel("Comprimento da pétala", fontsize=15)
eixo.grid(True)

# colocar média e desvio padrão
eixo.annotate('$media = {0:.2f}$\n$sigma = {1:.2f}$'.format(media, desvio_padrao),
             xy=(4.5,20), fontsize=20)

eixo.axvline(media, color= 'k', linestyle= '--')
eixo.annotate('média', xy=(media - 1.3, 28), fontsize= 20)

eixo.axvline(mediana, color= 'r', linestyle= '--')
eixo.annotate('mediana', xy=(mediana + 0.1, 31), fontsize= 20, color='r')
```

```python
# definição geral do ambiente gráfico
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

# média e desvio padrão
media = df.comprimento_pétala.mean()
desvio_padrao = df.comprimento_pétala.std()
mediana = df.comprimento_pétala.median()

# histograma
eixo.hist(df.comprimento_pétala, bins=20) # resultado normalizado
eixo.set_title("Histograma", fontsize=20, pad=10)
eixo.set_xlabel("Comprimento da pétala", fontsize=15)
eixo.grid(True)

# colocar média e desvio padrão
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(media, desvio_padrao), # mu = "mi" - usualmente média em estatística
             xy=(4.5,20), fontsize=20)                          # sigma - usualmente desvio padrão em estatística
 
eixo.axvline(media, color= 'k', linestyle= '--')
eixo.annotate('média', xy=(media - 1.3, 28), fontsize= 20)

eixo.axvline(mediana, color= 'r', linestyle= '--')
eixo.annotate('mediana', xy=(mediana + 0.1, 31), fontsize= 20, color='r')
```

```python
# definição geral do ambiente gráfico
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

df_iv = df[df.espécie == 'Iris-versicolor']

# média e desvio padrão
media = df_iv.comprimento_pétala.mean()
desvio_padrao = df_iv.comprimento_pétala.std()
mediana = df_iv.comprimento_pétala.median()

# histograma
eixo.hist(df_iv.comprimento_pétala, bins=20) # resultado normalizado
eixo.set_title("Íris Versicolor", fontsize=20, pad=10)
eixo.set_xlabel("Comprimento da pétala", fontsize=15)
eixo.grid(True)

# colocar média e desvio padrão
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(media, desvio_padrao), # mu = "mi" - usualmente média em estatística
             xy=(4.6,6), fontsize=20)                          # sigma - usualmente desvio padrão em estatística
 
eixo.axvline(media, color= 'k', linestyle= '--')
eixo.annotate('média', xy=(media - 0.5, 5.5), fontsize= 20)

eixo.axvline(mediana, color= 'r', linestyle= '--')
eixo.annotate('mediana', xy=(mediana - 0.7, 6.5), fontsize= 20, color='r')

fig.savefig('histograma_iv.png', bbox_inches='tight')
```

```python
from PIL import Image

largura, altura = Image.open('histograma_ivg.png').size # abrir uma delas para configurar o tamanho
combinada = Image.new('RGB', (3 * largura, altura))
espaco = 0

for imagem in map(Image.open, ['histograma_iv.png', 'histograma_is.png', 'histograma_ivg.png']):
    combinada.paste(imagem, (espaco, 0))
    espaco += largura
    
combinada.save('imagem_combinada.png')
```
